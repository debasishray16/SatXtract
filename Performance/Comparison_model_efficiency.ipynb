```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Set Seaborn theme for the plot
sns.set_theme(style="whitegrid")

# Data
epoch = [25, 50, 75, 100]
RESNET34 = [0.6991, 0.7796, 0.7375, 0]
RESNET50 = [0.7321, 0.8078,0.8074, 0]
RESNET101 = [0, 0.8008, 0, 0.7321]

# Plot setup
fig, ax = plt.subplots(figsize=(16, 9))
ax.set_facecolor('#ffffff')

# Plot each line with Seaborn
sns.lineplot(x=epoch, y=RESNET34, label='ResNet34', marker="o", color="blue", ax=ax)
sns.lineplot(x=epoch, y=RESNET50, label='ResNet50', marker="o", color="green", ax=ax)
sns.lineplot(x=epoch, y=RESNET101, label='ResNet101', marker="o", color="red", ax=ax)

# Customize plot appearance
plt.grid(True, linestyle='--', color='#BDBDBD')
plt.xticks(ticks=[25, 50, 75, 100])
plt.xlabel('Epoch')
plt.ylabel('Mean IOU')
plt.title('Epoch vs Mean IOU')

# Add annotations for each point
for x, y in zip(epoch, np.round(RESNET34, 4)):
    plt.text(x, y, f'({x},{y})', ha='right', fontsize=18)

for m, n in zip(epoch, np.round(RESNET50, 4)):
    plt.text(m, n, f'({m},{n})', ha='right', fontsize=18)

for p, q in zip(epoch, np.round(RESNET101, 4)):
    plt.text(p, q, f'({p},{q})', ha='right',fontsize=18)

plt.legend()
plt.show()
```

```python
import plotly.graph_objects as go
import numpy as np

# Data
epoch = [25, 50, 75, 100]
RESNET34 = [0.6991, 0.7796, 0.7375, 0]
RESNET50 = [0.7321, 0.8078, 0.8074, 0]
RESNET101 = [0, 0.8008, 0, 0.7321]

# Create interactive figure
fig = go.Figure()

# Add traces
fig.add_trace(go.Scatter(x=epoch, y=RESNET34, mode='lines+markers', name='ResNet34', line=dict(color='blue', width=3), marker=dict(size=10)))
fig.add_trace(go.Scatter(x=epoch, y=RESNET50, mode='lines+markers', name='ResNet50', line=dict(color='green', width=3), marker=dict(size=10)))
fig.add_trace(go.Scatter(x=epoch, y=RESNET101, mode='lines+markers', name='ResNet101', line=dict(color='red', width=3), marker=dict(size=10)))

# Add annotations dynamically
for x, y in zip(epoch, RESNET34):
    fig.add_annotation(x=x, y=y, text=f'({x},{y:.4f})', showarrow=True, arrowhead=2, font=dict(size=12, color='blue'))

for x, y in zip(epoch, RESNET50):
    fig.add_annotation(x=x, y=y, text=f'({x},{y:.4f})', showarrow=True, arrowhead=2, font=dict(size=12, color='green'))

for x, y in zip(epoch, RESNET101):
    fig.add_annotation(x=x, y=y, text=f'({x},{y:.4f})', showarrow=True, arrowhead=2, font=dict(size=12, color='yellow'))

# Customize layout
fig.update_layout(
    title='Epoch vs Mean IOU',
    xaxis_title='Epoch',
    yaxis_title='Mean IOU',
    template='plotly_dark',
    hovermode='x unified',
    font=dict(family='Arial, sans-serif', size=14, color='white'),
    legend=dict(title='Models', font=dict(size=14)),
    plot_bgcolor='#1e1e1e',
    paper_bgcolor='#1e1e1e',
    xaxis=dict(showgrid=True, gridwidth=1, gridcolor='gray'),
    yaxis=dict(showgrid=True, gridwidth=1, gridcolor='gray')
)

# Show interactive plot
fig.show()
```
